# Concatenar todos los archivos de la estacion meteorologica HOBO

```python
station = pd.concat([station_1, station_2], axis = 0)
station
```

```python
station.plot(subplots = True)
```

```python
station.info()
```

## Cambiar los segundos "02" a "00"

```python
# creamos nueva columna a partir del index
station['fecha'] = station.index
station
```

```python
# la nueva columna lo pasamos a formato str
station.fecha = station.fecha.astype(str)
station.info()
```

```python
#dividimos en dos nuevas columnas: date y hora

station[['date', 'hora']] = station.fecha.str.split(expand = True)
station
```

```python
# dividimos nuevamente la columna hora en tres comunas

station[['hora','minuto','segundo']] = station.hora.str.split(':', expand=True)
station
```

```python
# Ahora viene la magia
```

```python
station.segundo = 0 #la columna segundo lo cambiamos por valores igual a cero
station
station["tiempo"] = (station.date + " "+station.hora.astype("str") +
				":" + station.minuto.astype("str")+":"+station.segundo.astype("str"))
				
station.tiempo = pd.to_datetime(station.tiempo, format = '%Y-%m-%d %H:%M:%S')
station.set_index("tiempo", inplace = True)
station.drop(["fecha","date","hora","minuto","segundo"], axis = 1, inplace = True)

station
```

# Exportar a un solo archivo 

```python
station.to_csv('../data/002_processed/001_estacion_hobo/estacion_hobo.csv')
```
