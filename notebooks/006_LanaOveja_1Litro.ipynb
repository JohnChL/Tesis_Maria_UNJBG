# Importando estacion Hobo

```python
estacion_hobo = pd.read_csv('../data/002_processed/001_estacion_hobo/estacion_hobo.csv', index_col = 0, parse_dates = True)
hobo = estacion_hobo[['To','Ig','WS']]
hobo
```

# Concatenando las tres cocinas 1L de agua con tecnopor

```python
lanoveja_1l = pd.concat([df1, df2, df3, hobo], axis = 1)
lanoveja_1l
```

```python
## Eliminando valores fuera del rango

lanoveja_1l.loc[lanoveja_1l.T_p2 == -888.88, 'T_p2'] = float('NaN')
lanoveja_1l.loc[lanoveja_1l.T_m1 == 544.29, 'T_m1'] = float('NaN')
lanoveja_1l.loc[lanoveja_1l.T_m1 == 538.18, 'T_m1'] = float('NaN')

```
