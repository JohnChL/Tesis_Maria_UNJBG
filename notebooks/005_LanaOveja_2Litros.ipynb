# Importando estacion Hobo

```python
estacion_hobo = pd.read_csv('../data/002_processed/estacion_hobo.csv', index_col = 0, parse_dates = True)
hobo = estacion_hobo[['To','Ig','WS']]
hobo
```

# Concatenando las tres cocinas 1L de agua con tecnopor

```python
lanoveja_2l = pd.concat([df1, df2, df3, hobo], axis = 1)
lanoveja_2l
```

```python
## Eliminando valores fuera del rango

lanoveja_2l.loc[lanoveja_2l.T_p2 == -888.88, 'T_p2'] = float('NaN')
```

```python
lanoveja_2l
```

```python
lanoveja_2l = lanoveja_2l.loc['2024-01-19 00:00:00':'2024-01-21 23:59:00']
lanoveja_2l
```
